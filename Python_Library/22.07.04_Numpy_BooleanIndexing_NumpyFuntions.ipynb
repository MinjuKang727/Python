#### numpy 불리언 인덱싱
- 특정 조건에 맞으면 True, 아니면 False 값으로 배열을 만들어 조건에 맞는 값에 접근하는 방법

```python
import numpy as np
```

```python
# 50~ 99까지의 랜덤한 8개 값을 가지는 1차원 배열 생성
arr20 = np.array(np.random.randint(50,100, size=8))
print(arr20)
```

```python
# boolean(불리언) : True(참) 또는 False(거짓)으로 출력되는 자료형태
print(arr20 > 60)
print(arr20[arr20 > 60])
```

```python
# 짝수인 값들을 True로 출력시켜보세요~!
print(arr20 % 2 == 0)
```

```python
# 조건을 만족하는 배열의 원소 출력하기
# 배열명[열 내에서 찾고자 하는 조건식]
print(arr20[arr20 % 2 == 0])
```

```python
arr = np.array([[6, 9, 6, 8, 7], 
                [2, 3, 5, 4, 1]
               ])
```

```python
# arr에서 값이 홀수면 0으로 변경하는 코드를 작성해 보세요~!
# numpy 불리언 인덱싱 : 조건에 의해 인덱싱 된 배열의 요소에 원하는 값 대입******
arr[arr % 2 == 1] = 0
print(arr)
```

## 예제) BMI지수 및 과체중 BMI값 확인
- 데이터 로드

```python
data = np.loadtxt('data/height_weight.txt', delimiter=',') # delimiter : txt 파일 안에서 사용한 구분자
print(data)
```

### 데이터의 속성 확인
- 배열 형태
- 배열 차원수
- 배열 전체요소 개수

```python
print("배열 형태 :", data.shape)
print("배열 차원수 :", data.ndim)
print("배열 전체요소 개수 :", data.size)
```

#### 데이터에서 키와 몸무게 데이터들을 각각 변수에 담기

```python
# 키 
height = data[0]

# 몸무게
weight = data[1]
```

```python
print(height)
print(weight)
```

```python
bmi = weight / (height/100)**2
print(bmi)
```

#### BMI지수가 23이상 25이하인 사람은 몇 명일까요?
- 불리언 인덱싱에서는 
    - and → &
    - or → |
    - not → ~
- 를 쓴다.
- 그리고 두개 이상의 조건식을 쓸 때에는 각 조건식을 소괄호()로 묶어주어야 한다.

```python
# BMI가 23이상 25이하인 값 먼저 출력
# 출력된 값의 개수 구하는 명령 사용
print(len(bmi[(bmi >= 23) & (bmi <= 25)]))
```

## 예제) 영화 리뷰 평점 분석하기!

### 1. 데이터 로드

```python
movie = np.loadtxt('data/ratings.dat', delimiter="::",dtype=np.int64) # dtype : 데이터 타입을 속성값으로 변경
print(movie)
# 사용자번호(user_id), 영화번호(item_id), 영화평점(ratings), 입력시간(time)
```

### 2. 데이터 속성 확인
- 배열 형태
- 배열 차원수
- 배열 전체 요소 개수

```python
print("배열 형태 : ",movie.shape)
print("배열 차원수 : ",movie.ndim)
print("배열 전체 요소 개수 : ",movie.size)
```

### 3. 전체 영화의 평점 평균 구하기

```python
# 배열의 3번째 열에 접근하기
ratings = movie[:,2]
print(ratings)

# mean함수 사용하기
mean_ratings = np.mean(ratings)   # numpy 내장 함수

#전체 영화 평점 평균
print(mean_ratings)


mean_ratings2 = ratings.mean()    # 파이썬 내장 함수
print(mean_ratings2)
```

### 4. 사용자가 매긴 평점의 평균 구하기
- 사용자 번호의 유일한 값(유니크 값) 확인
- 사용자 번호의 유니크 값 개수 확인
- 1번 사용자의 데이터에 접근(불리언 인덱싱)
- 1번 사용자가 매긴 평점들의 평균 구하기

```python
# 평점을 매긴 사용자가 몇 명인지 알아보자!
user_id = np.unique(movie[:,0])  # .unique : 특정 배열의 유일한 값을 반환

# 사용자 번호의 유니크 값 확인
print(user_id)

# 사용자 번호의 유니크 값 개수 확인
print(len(user_id))
```

```python
# 1번 사용자의 데이터에 접근(불리언 인덱싱)
datas_1 = movie[movie[:,0] == 1]
print(datas_1)
```

```python
# 1번 사용자가 매긴 평점들의 평균 구하기
mean_ratings_1 = np.mean(datas_1[:,2])
print(mean_ratings_1)
```

- 1번 사용자는 전체 영화의 평점 평균보다 후하게 평점을 줬구나~!

### 5. 각 사용자가 매긴 평점들의 평균 구하기!

```python
print([[i,np.mean(movie[movie[:,0] == i][:,2])] for i in user_id])
```

```python
user_mean_list = []
for i in user_id :  # range(1, 6041, 1)
    
    # 1번 ~ 6040번 사용자가 매긴 평점들의 평균을 각각 구해주는 코드
    user_mean=[i,np.mean(movie[movie[:,0] == i][:,2])]
    
     # 빈 리스트에 사용자 번호와 그 사람이 매긴 평점의 평균을 리스트로 저장
    user_mean_list.append(user_mean)
print(user_mean_list)
```

### 6. 위에서 구한 각 사용자별 평점 평균이 4점 이상인 사용자만 구해보세요~!
- 사용자 번호와 명수를 구해봅시다!

```python
user_mean_arr = np.array(user_mean_list, dtype= "int64")
print(user_mean_arr)
```

```python
# 사용자별 평점 평균이 4점 이상인 사용자만 출력
user_mean_over4 = user_mean_arr[user_mean_arr[:,1] >= 4]
print(user_mean_over4[:,0])

# 명수 확인
print(len(user_mean_over4))
```

- 사용자가 매긴 평점의 평균이 4점 이상인 사람은 총 1544명

### 7. 10번 영화의 평점 평균 구하기!
- 직접 해보세요. ^-^

```python
# 10번 영화의 평점
ratings_of_10 = movie[movie[:, 1] == 10][:, [1,2]]
print(ratings_of_10)
```

### 8. 각 영화가 받은 평점의 연균 구하기

```python
print([[i,np.mean(movie[movie[:, 1] == i][:,2])] for i in np.unique(movie[:,1])])
```

```python
# 영화 번호 목록 
movie_id = np.unique(movie[:,1])
print(len(movie_id))
```

```python
movie_ratings = []

for i in movie_id :
    # 각 영화에 대한 평점의 평균 구해서 영화번호와 평점평균을 넣은 리스트를 영화 평점 리스트에 추가
    mean_rating_of_i = np.mean(movie[movie[:,1] == i][:,2])
    movie_ratings.append([i, mean_rating_of_i])
print(movie_ratings)
```
