- 요일별 사고 건수 시각화

```python
days = accident['요일'].value_counts()
print(days)
```

```python
# 요일 순서대로 변경
# Series에서 여러개의 인덱스에 접근해서 인덱스별로 원하는 순서대로 바꿔주기
days_accident = days[['월', '화', '수', '목', '금', '토', '일']]
print(days_accident)
```

#### 요일별 교통사고 건 수 시각화(Bar Graph)

```python
x = np.arange(7)
y = days_accident.values

plt.plot(x, y)
plt.xticks(np.arange(7), labels = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.show()

plt.bar(x, y)
# 문자로 tick 값들을 설정하기 위해서는 먼저 좌측부터 인덱스 번호로 개수를 설정해야 함
plt.xticks(np.arange(7), labels = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylim(480, 620)
plt.show()
```

```python
# 한글로 출력하기 위해서 폰트 지정하기!
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
plt.bar(x, y)
# 문자로 tick 값들ㅇ르 설정하기 위해서는 먼저 좌측부터 인덱스 번호로 개수를 설정해야 함
plt.xticks(np.arange(7), labels = days_accident.index)
plt.ylim(500, 620)

plt.show()
```

### 사고 유형이 차대차 사건 중에 죽거나 디친 사람이 많은 발생지시도를 알아보고 시각화해보자!

- 중요한 키워드들이 들어있는 위치 확인
    - 사고유형_대분류에서 '차대차'인 행만 출력
    - '차대차'인 행 중에서 '발생지시도'와 '사상자수' 칼럼 확인

```python
print(accident['사고유형_대분류'].value_counts())
```

```python
acci_cc = accident[accident['사고유형_대분류'] == '차대차']
print(acci_cc)
```

```python
print(acci_cc[['발생지시도','사상자수',]])
```

```python
# 사고유형이 '차대차'인 사건의 총 개수
print(len(acci_cc))
```

```python
# 발생한 '차대차' 사건 중 발생지시도별 사건 건수 확인
print(acci_cc['발생지시도'].value_counts())
```

#### 각 발생지시도 별로 사상자수의 합계 구하기

```python
acci_cc_loc = acci_cc[['발생지시도', '사상자수']].groupby('발생지시도').sum()
print(acci_cc_loc)
```

```python
print(acci_cc_loc.index)
print(len(acci_cc_loc.index))
acci_cc_loc.sort_values(by = '사상자수', ascending = False, inplace = True)
print(acci_cc_loc)
```

```python
plt.figure(figsize = (13, 7))

x = np.arange(17)
y = acci_cc_loc['사상자수']

plt.bar(x,y, color = 'orange')
plt.xticks(np.arange(17), labels = acci_cc_loc.index)

plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.title('차vs차 교통 사고의 사상자 수')

plt.show()
```

### 예제 4) 차vs차 교통사고 사상자 수 시각화하기
- x, y축 값 설정

```python
# x축에 들어갈 bar의 개수는 17개
plt.figure(figsize = (13, 7))

x = np.arange(len(acci_cc_loc.index))
y = acci_cc_loc['사상자수'].sort_values(ascending = False)

plt.bar(x, y)

plt.xticks(x, labels = acci_cc_loc.index, )

plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.title('차vs차 교통 사고의 사상자 수')

plt.show()
```

### 마지막 : 요일 별 야간 추돌 사고 건수를 구해봅시다!

```python
acci_2 = accident[(accident['사고유형_중분류'] == '추돌') & (accident['주야'] == '야간')]
acci_2
```

```python
acci_2_day = acci_2['요일'].value_counts()
acci_2_day = acci_2_day[['월', '화', '수', '목', '금', '토', '일']]
print(acci_2_day)
```

```python
plt.figure(figsize = (9, 6))
x = np.arange(7)
y = acci_2_day.values

plt.bar(x, y, edgecolor = 'red', linewidth = 2)
plt.ylim(0, 50)
plt.xticks(x, labels = acci_2_day.index)

plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.title('요일별 야간 차대사람 교통사고 건수')

plt.show()
```
