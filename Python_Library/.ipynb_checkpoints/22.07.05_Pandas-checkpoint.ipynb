```python
# index  
#    
#   
print(score_data.sort_index())
```

```python
# ascending : 
# ascending = False :  
print(score_data.sort_index(ascending = False))
```

```python
#   
# axis = 0 :   / axis = 1 :  
# drop() : axis = 0  drop/ axis = 1 :  drop
# sort_index() : axis = 0 (  )  / axis = 1 = (  ) 
print(score_data.sort_index(axis = 1))
print(score_data.sort_index(axis = 1, ascending = False))
```

- value  

```python
# DataFrame  index       
print(score_data.sort_values(by = '3')) # 3   
# asceding = False :  
print(score_data.sort_values(by = '3', ascending = False))
```

```python
# axis = 1 :  ()  
print(score_data.sort_values('', axis = 1, ascending = False))
```

### sum()
- axis   ,      
- .sum(axis = 0)       
- .sum(axis = 1)       

```python
print(score_data.sum()) #  axis = 0    
```

```python
print(score_data.sum(axis = 1))
```

-    DataFrame   ''  ~!

```python
score_data[''] = score_data.sum(axis = 1)
print(score_data)
```

### .mean()
-     DataFrame  ''  ~!

```python
score_data[''] = score_data.loc[:,'1':'4'].values.mean(axis = 1)
print(score_data)
```

```python
score_data[''] = score_data.iloc[:,:4].values.mean(axis = 1)
print(score_data)
```

```python
score_data[''] = score_data['']/4
print(score_data)
```

## min(), max()

```python
# 1   
print(score_data['1'].max())
```

```python
# axis = 0 :  () 
print(score_data[:].max())  # (default) : axis = 0
```

```python
#  () (,  X)
print(score_data.iloc[:, :4].max(axis = 1))
```

```python
#        ~!
print(score_data.iloc[:1,:4].max(axis = 1))
print(score_data.loc[:'',:'4'].max(axis = 1))
```

-            ~!

```python
sub_best = score_data.iloc[:,:4].max(axis = 1)
sub_worst = score_data.iloc[:,:4].min(axis = 1)
print(sub_best - sub_worst)
```

```python
sub_best = score_data.loc[:,:'4'].max(axis = 1)
sub_worst = score_data.loc[:,:'4'].min(axis = 1)
print(sub_best - sub_worst)
```
