```python
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time

addmissionRate = pd.DataFrame(columns = {'학년도', '학기', '모집시기', '모집학년', '지원학과', '모집인원', '지원현황(누계)', '경쟁률'})

# 1. 크롬 브라우저를 실행
driver = web.Chrome()

# 링크 주소로 이동
driver.get(f'https://haksa.knou.ac.kr/haksa/aee/appc/initEschAppcPrcn.do')

# 학년도 input 요소
# 학년도 입력 범위: 2011 ~ 2023
year = driver.find_element(By.CSS_SELECTOR, '#rcrtYosa')
# 학기 select 요소
semester = driver.find_element(By.CSS_SELECTOR, '#rcrtSeme')
semeOpt = driver.find_elements(By.CSS_SELECTOR, '#rcrtSeme > option')
# 모집시기 select요소
rcrt = driver.find_element(By.CSS_SELECTOR, '#rcrtEra')
rcrtOpt = driver.find_elements(By.CSS_SELECTOR, "#rcrtEra > option")
# 모집학년 select요소
grade = driver.find_element(By.CSS_SELECTOR, '#suptShgr')
gradeOpt = driver.find_elements(By.CSS_SELECTOR, "#suptShgr > option")
# 검색 button요소
search = driver.find_element(By.CSS_SELECTOR, 'button#btn_search')

for y in range(2011, 2024) :
    year.clear()
    year.send_keys(y)
    semester.click()
    for s in range(1, 3) :
        semeOpt[s].click()
        rcrt.click()
        for r in range(1, 3) :
            rcrtOpt[r].click()
            grade.click()
            for g in range(3) :
                gradeOpt[g].click()
                search.click()
                time.sleep(10)
                
                temp = pd.DataFrame(columns = list(addmissionRate.columns))
                idx = ['지원학과', '모집인원', '지원현황(오늘)', '지원현황(누계)', '경쟁률']
                for i in range(0, 5) :
                    if i % 5 == 2 :
                        continue
                    
                    datas = driver.find_elements(By.CSS_SELECTOR, f'#jqgrid .jqgrow td:nth-child({i+1})')
                    temp[idx[i]] = [d.text for d in datas]
                temp['학년도'] = y
                temp['학기'] = semeOpt[s].text
                temp['모집시기'] = rcrtOpt[r].text
                temp['모집학년'] = gradeOpt[g].text
                
                addmissionRate = pd.concat([addmissionRate, temp], axis=0)
                
driver.close()
addmissionRate.to_csv("입학지원율현황.csv")
```
