```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sort_value(x) :
    return x[1]
```

```python
import random as rd
# rd.randint() : 랜덤으로 정수를 반환하는 함수
# 예시) rd.randint(1, 10) : 1~10 까지의 랜덤 정수 반환


best_score = 0      # 최고 점수
leader_board = pd.DataFrame([], columns = ['플레이어', '점수'])
leader_board.index.names = ['순위']

while True :
    life = 3        # 목숨
    this_score = 0  # 이번판 점수

    # 게임 START!
    print('구구단을 외자! 구구단을 외자!')

    # 목숨이 남아있으면
    while life >= 0 :
        
        dan = rd.randint(2, 9)
        num = rd.randint(1, 9)

        while True:
            # 문제 내기 & 입력 받기
            ans = input(f'{dan} * {num} = ')  # 내 답
            correct = dan * num   # 정답

            # 정답이면,
            if str(correct) == ans:
                # 1점 획득
                this_score += 1
                print('정답!!')
                break

            # 오답이면
            else :
                # 목숨 1개 줆
                life -= 1
                print('땡! 틀렸습니다')

                # 아직 목숨이 남았으면,
                if life >= 0 :
                    print(f'남은 목슴 : {life}')

                # 최종 사망이면,   
                elif life == -1 :
                    player = input('플레이어 이름 : ')
                    leader_board.loc[len(leader_board)] = [player, this_score]
                    leader_board['점수'].sort_values()
                    
                    print('=================================')
                    print('순위\t플레이어\t점수')
                    print('---------------------------------')
                    
                    print(leader_board)
                    
                        
                            
                    # 최고 기록 경신이면,
                    if this_score <= best_score and len(leader_board) > 3:
                        li = np.arange(1, len(leader_board) + 1)
                        rank = max(li[(leader_board[1] == this_score) & (leader_board[0] == player)])
                        
                        print('---------------------------------')
                        print('내 성적')
                        print(f'{rank}위\t{player:<10s}\t{this_score:>10d}')
                        
                       
                    else :
                        print(f'최고 기록 경신 {this_score}점!!')
                        
                        # 최고 기록 reset
                        best_score = this_score
                        
                    break

    # 한판 끝나면 REGame 여부 질문
    play = input('목숨을 모두 소진하였습니다. 계속하시겠습니까? (Y|N) >> ')
    
    # Y 이면,
    if play == 'Y' :
        continue    # 한판 더

    # Y 아니면,
    else :
        break       # GG
```
