## 리스트(list) 란?
- 파이썬의 자료구조 형태중 하나
- 순서가 있는 수정가능한 객체의 집합
- 대괄호( [] )로 작성되어지며, 리스트 내부의 값은 콤마( , )으로 구분
- 추가, 수정, 삭제 가능

### 리스트명 = [요소1, 요소2, 요소3, ...]



## 인덱싱(indexing)
- 무엇인가를 ‘가리킨다‘는 의미 : 하나의 인덱스에 있는 값을 표시

## 슬라이싱(Slicing)
- 무엇인가를 ‘잘라낸다’는 의미 : 일정 범위의 인덱스에 있는 값을 표시

```python
# 리스트 인덱싱
# 리스트[인덱스] :인덱스에 위치한 값  반환
list1 = [2, 5, 7, 9, 10]
print(list1[0])
print(list1[3])
print(list1[2]+list1[-1])
```

```python
list2 = [1, 2, 3, ['a', 'b', 'c']] # 2차원 리스트
print(list2)
print(list2[3])
print(list2[3][1])
```

```python
# 리스트 슬라이싱
# 리스트[start 인덱스 : end 인덱스] : start 인덱스부터 end 인덱스 바로 전까지 값 반환(start <= x < end)
list3 = [0, 1, 2, 3, 4]
list3[1:3]
```

## 튜플(tuple) 이란?
- 파이썬의 자료구조 형태중 하나
- 순서가 있는 집합
- 소괄호( () )로 작성되어지며, 튜플의 내부 값은 콤마( , )으로 구분
- 추가, 수정, 삭제 불가능
- 동작속도가 빠르다.
- 정해져 있고 변하지 않는 값을 관리할 때 사용(변하는 값은 리스트로 관리)

### 튜플명 = (요소1, 요소2, 요소3, ...)

```python
a = ( )  # 빈튜플
b = (1, 2, 3) # 정수가 들어간 튜플
c = ('My', 'name', 'is', 'MJ') # 문자열 튜플
d = (1, 2, 'My', 'name') # 정수와 문자열이 들어간 튜플
e = (0, [1,2]) # 정수와 리스트가 들어간 튜플
f = (1, 2, ('My', 'name')) # 2차원 튜플
```

```python
# 튜플[인덱스] : 인덱스에 위치한 값 출력
tuple1 = (0, 1, 2, 3, ('a', 'b', 'c'), 5)
print(tuple1[2])
print(tuple1[4])
```

```python
# len(튜플) : 튜플의 값 개수 출력
tuple1 = (0, 1, 2, 3, ('a', 'b', 'c'), 5)
print(len(tuple1))
```

## 리스트(list)와 튜플(tuple) 공통점과 차이점
### 공통점
- 타입과 상관 없이 일련의 "요소(Element)"들을 가질 수 있다.
- 요소의 "순서"를 관리한다.

### 차이점
- 리스트는 "가변적(mutable)"이며, 튜플은 "불변적(immutable)"
- 리스트는 요소의 "개수가 명확하지 않고 유동적일 경우"에 사용
- 튜플은 요소 개수를 "사전에 정확히 알고 있고 변동이 없을 경우"에 사용
- (리스트에 비해 시스템 상 소모되는 메모리가 적고 속도가 빠름)

### in
- x in 문자열
- x in 리스트
- x in 튜플

### not in
- x not in 문자열
- x not in 리스트
- x not in 튜플

```python
str1 = "파이썬 최고"
```

```python
# in : 찾고자 하는 값(x)이 포함되어 있으면 True
"파이썬" in str1
```

```python
# not in : 찾고자 하는 값(x)이 포함되어 있지 않으면 True
"파이썬" not in str1
```

```python
list1 = [77, 38, 10]
33 in list1
```

```python
33 not in list1
```

```python
s = "Hi, My name is SeongWoo"
a = input("검색할 문자를 입력하세요 : ")
print(f'{a}는 {s.count(a)}번 들어있네유!' if a in s else f'{a}는 들어있지 않네유...')
```

```python
s = "Hi, My name is SeongWoo"
a = input("검색할 문자를 입력하세요 : ")
b = s.count(a)  # 검색 결과(동일 문자 개수)

# 문자열 중에서 검색할 문자가 들어있다면,
if a in s :
    print(f'{a}는 {s.count(a)}번 들어있네유!')
else :
    print(f'{a}는 들어있지 않네유...')
```
