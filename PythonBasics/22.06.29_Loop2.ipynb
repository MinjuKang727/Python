```python
# OX 퀴즈!
ox = input("OX입력 >> ")
score = 0
combo = 0

for i in ox :
    if i == 'O' :
        combo += 1
        score += combo
    else :
        combo = 0
print(f'{score}점')
```

```python
# 입력값을 X를 기준으로 자름 >> ox는 연속된 O 또는 ''이 들어있는 리스트
ox = input("OX입력 >> ").split('X')
score = 0

#연속된 O의 길이로 점수 추가
for i in ox :
    for j in range(len(i)) :
        # 점수 = (O의 인덱스) + 1 : 연속된 O 문자열에서 길이에 따른 추가점수 적용
        score += j + 1
        
print(f'{score}점')
```

```python
ox = input("OX입력 >> ")
# score리스트를 빈 리스트로 둘 경우 맨 처음 획득점수를 score[i] +1로 줄 시, 에러가 나기 때문에 0을 하나 추가
score = [0]

for i in range(len(ox)) :
    if ox[i] == 'O' :
        # O일 경우 획득 점수 : 이전 점수 + 1
        score.append(score[i] + 1)
    else :
        # X일 경우 획득 점수 : 0
        score.append(0)

# 획득 점수 : 정확히는 sum(score[1:])이지만, 0번째 요소가 0이기때문에 그냥 sum(score)해도 결과는 같음
print(f'{sum(score)}점')
```

```python
# OX 점수 구하기
# Hint OX를 입력받은 길이 만큼 range로 설정!

ox = input("OX입력 >> ")
score = 0
result = 0

for i in range(0, len(ox), 1) :
    if ox[i] == 'O' :
        score += 1        # 문제별 점수
        result += score   # result = result + score
    else :
        score = 0
print(f'{result}점')
```

```python
score_list = [['빅매일',89, 78, 99], ['김미희', 67, 88, 77], ['강예진', 100, 55, 96], ['최성우', 100, 100, 100]]
```

```python
# 성적표 제작!
print('='*50)
print('이름\t','Python\t', 'Java\t', 'DB\t', '평균\t', '학점')
print('='*50)

python, java, db = 0, 0, 0

for i in score_list :
    # 선생님별 평균
    avg = (i[1] + i[2] +i[3]) / 3
    #과목별 평균
    avg_P, avg_J, avg_D = 0, 0, 0

    # 평균에 대한 학점
    if avg >= 90 :
        grade = 'A'
    elif avg >= 80 :
        grade = 'B'
    elif avg >= 70 :
        grade = 'C'
    elif avg >= 60 :
        grade = 'D'
    else :
        grade = 'F'

    # 4 ~ 6번째 줄까지 출력
    print(i[0], '\t', i[1], '\t', i[2], '\t', i[3], '\t', int(avg), '\t', grade)

    # 과목별 평균 알고리즘
    python += i[1]   # python 점수의 합
    java += i[2]     # java 점수의 합
    db += i[3]       # db 점수의 합

avg_P = python / len(score_list)
avg_J = java / len(score_list)
avg_D = db / len(score_list)

print('='*50)

print('평균', '\t', int(avg_P), '\t', int(avg_J), '\t', int(avg_D))
```

### 네이버 영화 리뷰 간단 분석!

```python
# pickle : 리스트와 같은 자료구조 자체를 저장하거나 불러오게 해주는
#          라이브러리(기능들의 집합)
# 파이썬 자체의 기능이 아니라 외부의 기능을 불러오는 것
import pickle
```

```python
# 절대 경로: 처음부터 끝까지 경로를 다 쓰는 것
# 상대 경로: 내가 있는 파일에서부터 경로를 쓰는 것
# 만약 같은 파일 내에 있으면 그냥 'movie_review,pkl'만 써도 됨(반드시 확장자까지 써줘야함)
with open('data/movie_review.pkl','rb') as f : # rb: 읽기 전용/ as f :앞 부분을 그냥 f라고 부르겠다.
    my = pickle.load(f)
```

```python
# 0 : 부정리뷰, 1 : 긍정리뷰
my
```

```python
# 1. 전체 데이터 개수 확인
print(len(my))
```

```python
# 2. 리뷰만 출력
for i in my :
    print(i[0])
```

```python
# 3. 리뷰 길이 최대값 / 리뷰 내용
long = 0
for i in my :
    if len(i[0]) > long :
        long = len(i[0])
for i in my :
    if len(i[0]) == long :
        print(f"리뷰 최대길이는 '{long}'입니다.\n{i[0]}")
```

```python
long = max([len(i[0]) for i in my])
review = '\n'.join([i[0] for i in my if len(i[0]) == long])
print(f"리뷰 최대길이는 '{long}'입니다.\n{review}")
```

```python
# 4. 리뷰 길이 최소값 / 리뷰 내용
short = 144
for i in my :
    if len(i[0]) < short :
        short = len(i[0])
print(f"리뷰 최대길이는 '{short}'입니다.")
for i in my :
    if len(i[0]) == short :
        print(i[0])
```

```python
short = min([len(i[0]) for i in my])
review = '\n'.join([i[0] for i in my if len(i[0]) == short])
print(f"리뷰 최소길이는 '{short}'입니다.\n{review}")
```

```python
# 5. 리뷰 길이 평균값
avg_leng = 0
for i in my :
    avg_leng += len(i[0])
avg_leng /= len(my)
print(f"리뷰의 평균 길이는 '{avg_leng}'입니다.")
```

```python
avg_leng = sum([len(i[0]) for i in my])/len(my)
print(f"리뷰의 평균 길이는 '{avg_leng}'입니다.")
```

```python
# 6. 긍정리뷰/부정리뷰 개수

good = 0
bad = 0

for i in my :
    if i[1] == 1 :
        good += 1
    else :
        bad += 1
        
print(f"긍정 '{good}'개")
print(f"부정 '{bad}'개")
```

```python
good = sum([j for i,j in my])
bad = len(my) - good
print(f"긍정 '{good}'개")
print(f"부정 '{bad}'개")
```

```python
search = input("검색할 데이터 입력 >> ")
count = 0
for i, j in my :
    if search in i :
        count += 1
print(f"'{search}'(이)가 포함된 리뷰는 '{count}'개 입니다")
```

```python
# 두번째 방법
re_list = []

# 0 또는 1 값을 re_list에 저장하는 반복문
for i in my :
    re_list.append(i[1])

neg = re_list.count(0)
pos = re_list.count(1)

print(f"긍정 '{pos}'개")
print(f"부정 '{neg}'개")
```
