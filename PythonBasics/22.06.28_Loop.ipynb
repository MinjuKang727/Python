### break
- 반복문 밖으로 나가는 명령

```python
# While True(무한루프) - break 예제
while True:
    print('무한 루프') # 조건문이 True이기 때문에 계속 '무한 루프' 출력~~~~~
    break             # break를 써서 반복문을 빠져나가자! 
```

```python
                    # while 조건문 :  으로 써도 되지만 조건이 명확하지 않을 시,
number = 1
while True:        # while True : 로 쓰고 
    print(number)
    number += 1
    if number > 3 : # while문 안에 if문을 써서 조건 달성시,
        break       # break로 반복문을 빠져나가도록 코딩을 해도 됨
```

### continue
- 반복문 시작으로 돌아가는 명령

```python
# while - continue 예제
num = 1
while num <= 10 :
    print(f'{num}번째 반복중..')
    if num >= 4 :
        print('여기서 부터는 num이 증가하지 않습니당..')
        #continue
        break
    num += 1
```

```python
score_list = [['빅매일',89, 78, 99], ['김미희', 67, 88, 77], ['강예진', 100, 55, 96]]

print('='*49)

menu = ['이름', 'Python', 'Java', 'DB', '평균', '학점']
print('\t'.join(menu))

print('='*49)

python = 0
java = 0
db = 0
# score_list의 각 요소 : 선생님 개별 성적표
for i in score_list:
    # (평균) = (성적표의 성적의 합) / (과목 수)
    avg = int(sum(i[1:])/3)
    #개별 성적표에 평균 추가
    i.append(avg)
    #평균 60점 미만 : F등급
    if avg < 60 :
        i.append('F')
    #평균 60점 이상 70점 미만 : D등급
    elif avg < 70 :
        i.append('D')
    #평균 70점 이상 80점 미만 : C등급
    elif avg < 80 :
        i.append('C')
    #평균 80점 이상 90점 미만 : B등급
    elif avg < 90 :
        i.append('B')
    #평균 90점 이상 : A등급
    else :
        i.append('A')
    # (각 과목별 총점) += (개인 성적)
    python += i[1]
    java += i[2]
    db += i[3]
    # 개별 이름, 성적표 출력
    for j in i :
        print(j, end = '\t')
    print('')
# (과목별 평균) = (과목별 개인 성적 총합) / (시험 응시자 수)  
python /= len(score_list)
java /= len(score_list)
db /= len(score_list)

avg = ['평균', int(python), int(java), int(db)]

print('='*49)
#과목별 평균 출력
for i in avg:
    print(i, end = '\t')
```

```python
score_list = [['빅매일',89, 78, 99], ['김미희', 67, 88, 77], ['강예진', 100, 55, 96]]

print('='*49)

menu = ['이름', 'Python', 'Java', 'DB', '평균', '학점']
print('\t'.join(menu))

print('='*49)

avg_list = ['평균']
#과목 갯수만큼 0으로 평균 set
for i in range(len(menu)-3):
    avg_list.append(0)
    
# score_list의 각 요소 : 선생님 개별 성적표
for i in score_list:
    avg = 0
    # 성적부분의 인덱스 범위에서 반복문 실행
    for j in range(1, len(i)):
        # 개인 평균에 각 과목 점수 더하기
        avg += i[j]
        # 과목별 평균에 각 과목 점수 더하기
        avg_list[j] += i[j]
    #개인 평균 구하기
    avg = int(avg / len(i[1:]))
    #개별 성적표에 평균 추가
    i.append(avg)
    
    #선생님 등급 구하기
    #평균 60점 미만 : F등급
    if avg < 60 :
        i.append('F')
    #평균 60점 이상 70점 미만 : D등급
    elif avg < 70 :
        i.append('D')
    #평균 70점 이상 80점 미만 : C등급
    elif avg < 80 :
        i.append('C')
    #평균 80점 이상 90점 미만 : B등급
    elif avg < 90 :
        i.append('B')
    #평균 90점 이상 : A등급
    else :
        i.append('A')   
    
    # 개별 이름, 성적표 출력
    for j in i:
        print(j, end = '\t')
    print('')
    
# (과목별 평균) = (해당 과목의 성적 총합) / (시험 응시자 수)
for i in range(1, len(avg_list)) :
    avg_list[i] = int(avg_list[i] / len(score_list))

print('='*49)
#과목별 평균 출력
for i in avg_list :
    print(i, end = '\t')
```
