### 치환
- A는 양 손에 빨간색 펜과 파란색 펜을 하나씩 들고 있다.
- A가 양 손을 동시에 쓰지 않고 빨간색 펜과 파란색 펜을 바꿔서 들기 위해서는 어떻게 해야 할까?

##### 한 손에 든 펜을 놓고 들고 있던 펜을 다른 손으로 바꿔 잡고 내려놓았던 펜을 남은 손으로 집어 든다.


```python
# 변수 a와 b의 값을 치환하시오.
a = 3
b = 7

print(("a : {}".format(a)))
print(("b : {}".format(b)))
```

```python
temp = a
a = b
b = temp

print(("a : {}".format(a)))
print(("b : {}".format(b)))
```

```python
# 간편한 치환 방법!!
a,b = b, a

print(("a : {}".format(a)))
print(("b : {}".format(b)))
```

```python
# 간편한 대입 방법
a = b = c = 3
print(a, b, c)
```
