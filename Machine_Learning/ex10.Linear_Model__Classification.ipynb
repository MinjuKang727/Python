

- 회귀용 선형 모델 : 선형 함수가 바로 결정 값이 됨
- 분류용 선형 모델 : 선형 함수가 결정 경계의 역할을 함
    - Logistic Regression(Regression 단어가 붙지만 분류용 모델) : 딥러닝으로 이어짐
    - Linear Support Vector Machine : 컴퓨터가 좋아져서 Logistic Regression을 더 많이 사용


- 단순 선형 모델로는 : 분류가 제대로 되지 않음
    - 타겟의 범위가 점점 커지면 제대로 결정하지 못하는 타겟들이 등장하게 됨

- Logistic 회귀는 sigmoid 함수를 쓰기 때문에 값이 0과 1 사이의 값으로 나옴
- 양수 값 : 내가 관심 갖는 값
- 선형 회귀 모델 : Data >> 선형 모델에 넣음 : y와 y를 예측한 값 사이의 오차(MSE를 사용)를 함수로 그린 것 : Loss, Cost함수
    - Optimizer(최적화 도구)로 경사하강법(GD, Gradient Descent)를 사용하여 오차를 최소화
- 분류 선형 회귀 : 특성 상, MSE를 오차로 사용할 경우 Optimizer(최적화) 기준으로 사용할 경우 별로 오차가 줄어들지 않음(0 or 1)
    - Binary Cross Entropy로 정확도 판단


- 실제 정답이 1이면 왼쪽 log함수 값, 실제 정답이 0일 경우, 오른쪽 결과 함수를 고르게 됨


- Support vector : 가장자리에 있는 데이터
- 결정 경계가 데이터와 너무 가까워지지 않게 margin을 줌


- 선형 회귀 모델은 규제 방법을 클래스로 분류해 둠
    - L1 선형 회귀 모델의 규제 방법 : Lasso - 모든 가중치를 0으로 깍아버림 >> 특성 선택이 일어난다.
    - L2 선형 회귀 모델의 규제 방법 : Ridge
    - 규제 크기 파라미터 : alpha (값이 클 수 록 규제가 강해짐)
- 규제 크기 커짐 : 입력 특성의 가중치를 크게줌(영향력 커짐) : 오차 변화율도 커지게 되므로 경사하강법을 쓰게 되면 변화 폭이 커짐
- 선형 분류 모델: 파라미터로 규제 방법을 결정
    - 규제 방법이 따로 없고 규제 파라미터로 조절 : C(대문자 c) (값이 클수록 규제가 약해진다.: 계산 과정에 역수를 취하기 때문)


- 선형 모델 : 희소데이터에서 잘 작동
- 희소 데이터 : 텍스트 데이터
