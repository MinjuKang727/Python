# 5. 핑퐁 게임(Ping Pong Game)

## 5.1 핑퐁 게임 규칙
- 네모난 공간에 플레이어와 컴퓨터의 라켓이 있고 공이 주어진다.
- 플레이어는 라켓을 움직여서 공을 친다.
- 컴퓨터는 공을 따라서 오는 공을 친다.
- 상대방이 공을 놓치게 되면 점수를 얻는다.
- 점수를 10점 얻는 사람이 이긴다.
