## 3.6 공놀이 하기

```python
## 공놀이 초기 설정 예제

import pygame

# 게임 스크린 크기
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

# 색 정의
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

# pygame 초기화
pygame.init()

# 윈도우 제목
pygame.display.set_caption("Ball")

# 스크린 정의
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# 게임 화면 업데이트 속도
clock = pygame.time.Clock()


## 공 변수 정의 및 초기화 예제

# 공 초기 위치, 크기, 속도
ball_x = int(SCREEN_WIDTH / 2)
ball_y = int(SCREEN_HEIGHT / 2)
ball_dx = 4  # 방향 값
ball_dy = 4  # 방향 값
ball_size = 40


## 게임 반복 구간 예제

# 게임 종료 전까지 반복
done = False

# 게임 반복 구간
while not done:
    # 이벤트 반복 구간
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
            
    ## 게임 로직 구간
    
    # 속도에 따라 원형 위치 변경
    ball_x += ball_dx
    ball_y += ball_dy
    
    # 공이 스크린을 벗어날 경우
    if (ball_x + ball_size) > SCREEN_WIDTH or (ball_x - ball_size) < 0:
        ball_dx = ball_dx * -1
    if (ball_y + ball_size) > SCREEN_HEIGHT or (ball_y - ball_size) < 0:
        ball_dy = ball_dy * -1
        
    ## 공놀이 게임 화면 그리기 예제
    
    # 스크린 채우기
    screen.fill(WHITE)
    
    # 화면 그리기 구간
    # 공 그리기
    pygame.draw.circle(screen, BLUE, [ball_x, ball_y], ball_size, 0)
    
    # 화면 업데이트
    pygame.display.flip()
    
    # 초당 60 프레임으로 업데이트
    clock.tick(60)
    
# 게임 종료
pygame.quit()
```

## 이미지 사용

- 이미지 불러오기<br>
<b>pygame.image.load()</b>

```python
## 이미지 사용 윈도우 설정 예제

import pygame
import os

# 게임 스크린 크기
SCREEN_WIDTH = 640
SCREEN_HEIGHT = 320

# 색 정의
LAND = (160, 120, 40)

# pygame 초기화
pygame.init()

# 윈도우 제목
pygame.display.set_caption("Image")

# 스크린 정의
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# 게임 화면 업데이트 속도
clock = pygame.time.Clock()


## 이미지 로드 예제

# image 경로 설정
current_path = os.path.dirname('pygame.py')
image_path = os.path.join(current_path, 'image')

# 배경 이미지 로드
background_image = pygame.image.load(os.path.join(image_path, 'plains.png'))

# 이미지 로드
mushroom_image_1 = pygame.image.load(os.path.join(image_path, 'mushroom1.png'))
mushroom_image_2 = pygame.image.load(os.path.join(image_path, 'mushroom2.png'))
mushroom_image_3 = pygame.image.load(os.path.join(image_path, 'mushroom3.png'))


## 이미지 사용 예제

# 게임 종료 전까지 반복
done = False

# 게임 반복 구간
while not done:
    # 이벤트 반복 구간
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
            
    # 게임 로직 구간
    
    # 화면 삭제 구간
    
    # 스크린 채우기
    screen.fill(LAND)
    
    
    # 화면 그리기 구간
    # 배경 이미지 그리기
    screen.blit(background_image, background_image.get_rect())
    
    
    # 버섯 이미지 그리기
    screen.blit(mushroom_image_1, [100, 250])
    screen.blit(mushroom_image_2, [300, 240])
    screen.blit(mushroom_image_3, [450, 270])
    
    # 화면 업데이트
    pygame.display.flip()
    
    # 초당 60 프레임으로 업데이트
    clock.tick(60)
    
# 게임 종료
pygame.quit()
```

## 키보드 조작

```python
## 키보드 조작 윈도우 설정 예제

import pygame
import os

# 게임 스크린 크기
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600


# 색 정의
GRAY = (200, 200, 200)


# pygame 초기화
pygame.init()


# 윈도우 제목
pygame.display.set_caption("Keyboard")


# 스크린 정의
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))


# 게임 화면 업데이트 속도
clock = pygame.time.Clock()


## 키보드 이미지 초기 설정 예제

# image 경로 설정
current_path = os.path.dirname("pygame.py")
image_path = os.path.join(current_path, 'image')

# 키보드 이미지 초기 설정
keyboard_image = pygame.image.load(os.path.join(image_path, 'keyboard.png'))
keyboard_x = int(SCREEN_WIDTH / 2)  # 키보드 이미지의 좌상단 x좌표
keyboard_y = int(SCREEN_HEIGHT / 2)  # 키보드 이미지의 좌상단 y좌표
keyboard_dx = 0  # 키보드 방향 x좌표
keyboard_dy = 0  # 키보드 방향 y좌표


## 키보드 조작 예제

# 게임 종료 전까지 반복
done = False

# 게임 반복 구간
while not done:
    # 이벤트 반복 구간
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
        # 키가 눌릴 경우
        elif event.type == pygame.KEYDOWN:
            if event.type == pygame.K_LEFT:
                keyboard_dx = -3
            elif event.key == pygame.K_RIGHT:
                keyboard_dx = 3
            elif event.key == pygame.K_UP:
                keyboard_dy = -3
            elif event.key == pygame.K_DOWN:
                keyboard_dy = 3
        # 키가 놓일 경우
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                keyboard_dx = 0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                keyboard_dy = 0
       
    
    ## 키보드 이미지 위치 및 그리기 예제
    # 게임 로직 구간
    # 키보드 이미지의 위치 변경
    keyboard_x += keyboard_dx
    keyboard_y += keyboard_dy
    
    # 스크린 채우기
    screen.fill(GRAY)
    
    # 화면 그리기 구간
    # 키보드 이미지 그리기
    screen.blit(keyboard_image, [keyboard_x, keyboard_y])
    
    # 화면 업데이트
    pygame.display.flip()
    
    # 초당 60 프레임으로 업데이트
    clock.tick(60)
    
# 게임 종료
pygame.quit()
```

```python
## 마우스 조작 윈도우 설정 예제

import pygame
import os

# 게임 스크린 크기
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

# 색 정의
LIGHT_GREEN = (200, 255, 50)

# pygame 초기화
pygame.init()

# 윈도우 제목
pygame.display.set_caption("Click mouse")

# 스크린 정의
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# 게임 화면 업데이트 속도
clock = pygame.time.Clock()


## 마우스 이미지 정의 예제

# image 경로 설정
current_path = os.path.dirname("pygame.py")
image_path = os.path.join(current_path, 'image')

# 마우스 이미지 초기 설정
mouse_image = pygame.image.load(os.path.join(image_path,'mouse.png'))
mouse_x = int(SCREEN_WIDTH / 2)  # 마우스 이미지의 좌상단 x좌표
mouse_y = int(SCREEN_HEIGHT / 2)  # 마우스 이미지의 좌상단 y좌표


## 마우스 커서 숨기기 예제

# 마우스 커서 숨기기
pygame.mouse.set_visible(False)


## 마우스 위치 값 예제

# 게임 종료 전까지 반복
done = False

# 게임 반복 구간
while not done:
    # 이벤트 반복 구간
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
        # 마우스 버튼이 눌릴 경우    
        elif event.type == pygame.MOUSEBUTTONDOWN :
            if event.button == 1:
                mouse_image = pygame.image.load(os.path.join(image_path,'mouse_left_click.png'))
            elif event.button == 3:
                mouse_image = pygame.image.load(os.path.join(image_path,'mouse_right_click.png'))
        # 마우스 버튼이 놓일 경우
        elif event.type == pygame.MOUSEBUTTONUP:
            mouse_image = pygame.image.load(os.path.join(image_path,'mouse.png'))
    
    # 게임 로직 구간
    
    
    # 마우스 위치 값 가져오기
    pos = pygame.mouse.get_pos()
    mouse_x = pos[0]
    mouse_y = pos[1]
    
    
    ## 마우스 조작 화면 그리기 예제
    
    # 스크린 화면 채우기
    screen.fill(LIGHT_GREEN)
    
    # 화면 그리기 구간
    # 마우스 이미지 그리기
    screen.blit(mouse_image, [mouse_x, mouse_y])
    
    
    # 화면 업데이트
    pygame.display.flip()
    
    # 초당 60 프레임으로 업데이트
    clock.tick(60)
    
# 게임 종료
pygame.quit()
```

## 3.10 마우스 및 사운드 사용

```python
## 사운드 사용 윈도우 설정 예제
import pygame
import os

# 게임 스크린 크기
SCREEN_WIDTH = 640
SCREEN_HEIGHT = 400

# 색 정의
BLACK = (0, 0, 0)

# pygame 초기화
pygame.init()

# 윈도우 제목
pygame.display.set_caption("Sound")

# 스크린 정의
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# 게임 화면 업데이트 속도
clock = pygame.time.Clock()


## 배경 이미지, 배경 음악, 효과음 로드 예제

# image 경로 설정
current_path = os.path.dirname("pygame.py")
image_path = os.path.join(current_path, 'image')

# 배경 이미지 로드
background_image = pygame.image.load(os.path.join(image_path, 'equalizer.png'))

# 배경 음악 로드
pygame.mixer.music.load(os.path.join(image_path, 'bgm.wav'))
pygame.mixer.music.play(-1)  # 무한 반복 재상

# 효과음 로드
sound = pygame.mixer.Sound(os.path.join(image_path, 'sound.wav'))

# 게임 종료 전까지 반복
done = False

# 게임 반복 구간
while not done:
    # 이벤트 반복 구간
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
        if event.type == pygame.MOUSEBUTTONDOWN:
            sound.play()  # 효과음 재생
       
    
    ## 게임 화면 구성 예제
    
    # 게임 로직 구간
    
    # 화면 삭제 구간
    
    # 스크린 채우기
    screen.fill(BLACK)
    
    # 화면 그리기 구간
    # 배경 이미지 그리기
    # screen.blit(background_image, background_image.get_rect())
    # 이미지가 너무 왼쪽 상단으로 치우쳐서 예제와 다르게 좌표 바꿈
    # get_rect() = <rect(좌측상단 x좌표, 좌측상단 y좌표, 폭, 높이)>
    screen.blit(background_image, [(SCREEN_WIDTH - background_image.get_rect()[-2]) / 2, (SCREEN_HEIGHT - background_image.get_rect()[-1]) / 2])
    
    
    # 화면 업데이트
    pygame.display.flip()
    
    # 초당 60 프레임으로 업데이트
    clock.tick(60)
    
# 게임 종료
pygame.quit()
```
