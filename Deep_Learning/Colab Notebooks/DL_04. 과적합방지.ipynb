## 센서 데이터 불러오기

```python
import pandas as pd

sonar = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep_Learning/data/sonar.csv', header=None)
sonar.head()
```

```python
sonar.shape
```

```python
# 데이터 분리
datasets = sonar.values

X = datasets[:, :-1]
y = datasets[:, -1]
```

```python
# 문자열로 된 데이터를 숫자로 변경
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(y)
y = e.transform(y)
y
```

```python
# 훈련과 테스트셋으로 분리, 데이터 섞어주기
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout  # Dropout : 과대적합 예방 (랜덤하게 노드 사용을 막음)

# 모델 설계
# 신경망 생성
model1 = Sequential()

# 입력층
model1.add(Dense(units=60, input_dim=60, activation="relu"))

# 중간층
model1.add(Dense(units=120, activation="relu"))
model1.add(Dense(units=60, activation="relu"))
model1.add(Dense(units=30, activation="relu"))

# 출력층
model1.add(Dense(units=1, activation="sigmoid"))

model1.summary()
```

```python
# 실행 방법 설정
model1.compile(loss="binary_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
import numpy as np

X_train = X_train.astype(np.float32)
```

```python
# 학습
model1.fit(X_train, y_train, epochs=200)
```

```python
X_test = X_test.astype(np.float32)

model1.evaluate(X_train, y_train)
```

```python
model1.evaluate(X_test, y_test)
```

### validation_split
- 과대적합 확인

```python
# 학습
# validation : 학습 데이터의 일정비율을 과적합 확인용으로 사용
model1.fit(X_train, y_train, epochs=1000, validation_split=0.3)
```

```python
X_test = X_test.astype(np.float32)

model1.evaluate(X_train, y_train)
```

### Dropout
- 과대적합 예방

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout  # Dropout : 과대적합 예방 (랜덤하게 노드 사용을 막음)

# 모델 설계
# 신경망 생성
model1 = Sequential()

# 입력층
model1.add(Dense(units=64, input_dim=60, activation="relu"))

# 중간층
model1.add(Dense(units=128, activation="relu"))
model1.add(Dropout(0.6))
model1.add(Dense(units=64, activation="relu"))
model1.add(Dropout(0.6))
model1.add(Dense(units=32, activation="relu"))
model1.add(Dropout(0.6))

# 출력층
model1.add(Dense(units=1, activation="sigmoid"))

model1.summary()
```

```python
# 실행 방법 설정
model1.compile(loss="binary_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
# 학습
# validation : 학습 데이터의 일정비율을 과적합 확인용으로 사용
model1.fit(X_train, y_train, epochs=1000, validation_split=0.3)
```

```python
X_test = X_test.astype(np.float32)

model1.evaluate(X_train, y_train)
```

```python
model1.evaluate(X_test, y_test)
```

### 이미지를 학습하지 않는 이유
- MLP, CNN : 딥러닝 모델
  - MLP > 이미지 학습은 가능 > 판단은 엄청 못함
    - 위치를 기반으로 학습 >> 위치가 조금만 달라져도 다른 값으로 파악
    - 크기도 중요한 변수 >> 이미지들의 크기가 거의 다 같아야 함
    - 이미지의 모양은 학습 가능
    
  - CNN : 이미지를 학습할 수 있음 > 판단을 잘함
- openCV : 라이브러리
  - openCV(라이브러리) : 이미지를 다루는 방법
