# 활성화 함수

- 이전 층의 값을 다음으로 비선형 변환을 해서 전달하는 함수
  - 실제 뉴런의 구조가 비선형으로 동작하기 때문에
  - 경험적으로 비선형 변환을 쓰는 것이 신경망 동작이 더 잘 되었음

- 종류 : sigmoid(입력, 은닉, 출력), tanh, relu(입력, 은닉), softmax(출력) 등

# sigmoid 함수 만들기
<img src="https://drive.google.com/uc?id=1Uay2cYaP7JHr6MOEmJosuJVgFLSzT_rI">

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_func(x) :
  return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid_func(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.grid()
```

### Sigmoid 미분
<img src="https://drive.google.com/uc?id=1M5d0Qu3thFYowEUm-dRuh7TX9bBBQs65">

```python
def d_sigmoid_func(x) :
  return sigmoid_func(x) * (1 - sigmoid_func(x))

x = np.arange(-5.0, 5.0, 0.1)
y = d_sigmoid_func(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.grid()

# 최대값이 0.25
# 오차역전파 시, 활성화 함수를 미분해서 곱하는데 층이 많아지면 활성화 함수의 곱이 0에 가까워져서 파라미터를 업데이트하지 못하는 문제가 발생 : vanishing gradient
```

# tanh 함수

- sigmoid의 문제점을 해결하기 위한 대안

```python
# tanh 함수
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.grid()
# -1부터 1까지의 범위 값을 가진다.
```

### tanh 미분
<img src="https://drive.google.com/uc?id=1riMrWEj8smZx2Nl-x62YMsoPsYfCJdv9">

```python
def d_tanh(x) :
  return 1 - np.tanh(x) * np.tanh(x)

x = np.arange(-5.0, 5.0, 0.1)
y = d_tanh(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.grid()

# 최대값이 1이므로 sigmoid에 비해서 더 많이 곱할 수 있음
# But, tanh 역시 층이 깊어지면 0에 가까워짐 : vanishing gradient
```

# relu 함수

- 0보다 작으면 0이고 0보다 크면 x인 함수

```python
def relu_fun(x) :
  # y = 0 (x < 0) 
  # y = x (x >= 0)
  # 0과 x 중에서 큰 값을 반환
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu_fun(x)

plt.plot(x, y)
plt.grid()
```
