## Melspectrum
- 소리 데이터를 샘플링 할 때,
  - 저주파 부분(사람이 잘 듣는 주파수 대역)은 많이 샘플링 --> 간격을 짧게 함
  - 고주파 부분(사람이 잘 못 듣는 주파수 대역)은 적게 샘플링 --> 간격을 길게 함
- 사람의 귀가 듣는 특성을 반영
- 소리 데이터를 이미지 데이터로 변환할 수 있음

## MFCC
- Melspectrum을 디지털로 변환한 것

```python
# n_mels : 멜 스펙트럼의 수 (세로축의 눈금 수)
# 시간별 소리의 크기, 주파수 성분을 시각화
S = librosa.feature.melspectrogram(wav, sr=sr, n_mels=128)


# 시각화
log_S = librosa.power_to_db(S, ref = np.max)
librosa.display.specshow(log_S, sr=sr, x_axis="time", y_axis="mel")

# y축은 주파수(위: 높은 음, 아래: 낮은 음)
```

## 섬집아기

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display

# 불러올 소리 데이터가 있는 폴더로 경로 설정
data_dir = "/content/drive/MyDrive/Colab Notebooks/Deep_Learning/data/"

# 소리 데이터 읽기
# wav : 소리 데이터
# sr : sampling rate (=초당 데이터 수)
wav, sr = librosa.load(data_dir + "섬집아기.wav")

# 소리 데이터 정보 출력
print(sr)
print(wav.shape)
print(wav.shape[0]/ float(sr))

# 총 67253개의 데이터로 구성되어 있고 초당 22050개의 데이터로 구성
```

```python
# n_mels : 멜 스펙트럼의 수 (세로축의 눈금 수)
# 시간별 소리의 크기, 주파수 성분을 시각화
S = librosa.feature.melspectrogram(wav, sr=sr, n_mels=128)


# 시각화
log_S = librosa.power_to_db(S, ref = np.max)
librosa.display.specshow(log_S, sr=sr, x_axis="time", y_axis="mel")

# y축은 주파수(위: 높은 음, 아래: 낮은 음)
```

```python
# 사운드 데이터를 동일한 길이로 만들어주는 함수

def pad1d(wav, i) :  # pad1d(사운드 데이터, 길이)
  # 사운드 데이터가 i보다 긴 경우
  if wav.shape[0] > i :
    return wav[:i]

  # 사운드 데이터가 i보다 짧은 경우
  else :
    return np.hstack(np.zeros(i-wav.shape[0]))

# MFCC 데이터(2차원)를 동일한 크기로 만들어주는 함수
def pad2d(wav, i) :
  if wav.shape[1] > i :
    return wav[:, :i]
  else :
    return np.hstack(wav, np.zeros(wav.shape[0], i - wav.shape[1]))
```

## 폴더에 있는 사운드 데이터를 특성 데이터와 라벨 데이터로 저장

```python
import os

data_dir = "/content/drive/MyDrive/Colab Notebooks/Deep_Learning/data/baby_sound/"

train_dir = data_dir + "train/"

# train폴더의 사운드 파일을 읽어서 wav를 특성데이터로 저장하고 파일명을 라벨데이터로 저장
for fname in os.listdir(train_dir) :
  # wav 파일일 아닌 경우 아무런 처리도 하지 않음
  if ".wav" not in fname :
    continue

  # 파일 이름을 라벨데이터에 저장(crying, laugh, silence)
  # 파일 이름 : 울음종류 (인덱스).wav
  label = fname.split(" ")[0]
  y_train.append(label)

  # 사운드 데이터를 같은 길이로 만들어서 특성 데이터로 저장
  wav, sr = librosa.load(train_dir + fname)

  pad_x = pad1d(wav, 30000)
  X_train.append(pad_x)
```

```python
import os

data_dir = "/content/drive/MyDrive/Colab Notebooks/Deep_Learning/data/baby_sound/"

test_dir = data_dir + "test/"

# test폴더의 사운드 파일을 읽어서 wav를 특성데이터로 저장하고 파일명을 라벨데이터로 저장
for fname in os.listdir(test_dir) :
  # wav 파일일 아닌 경우 아무런 처리도 하지 않음
  if ".wav" not in fname :
    continue

  # 파일 이름을 라벨데이터에 저장(crying, laugh, silence)
  # 파일 이름 : 울음종류 (인덱스).wav
  label = fname.split(" ")[0]
  y_test.append(label)

  # 사운드 데이터를 같은 길이로 만들어서 특성 데이터로 저장
  wav, sr = librosa.load(test_dir + fname)

  pad_x = pad1d(wav, 30000)
  X_test.append(pad_x)
```

```python
# 리스트로 된 데이터를 배열의 행 데이터로 변환
import numpy as np
X_train = np.vstack(X_train)
X_test = np.vstack(X_test)

X_train.shape, X_test.shape
```

```python
# y를 원핫 인코딩
import pandas as pd

y_train_en = pd.get_dummies(y_train)
y_test_en = pd.get_dummies(y_test)

y_train_en.shape, y_test_en.shape
```

## 1. 사운드 데이터를 DNN으로 학습하기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model1 = Sequential()

# 입력층
model1.add(Dense(units=512, input_shape=(30000,), activation="relu"))
model1.add(Dropout(0.3))

# 은닉층
model1.add(Dense(units=256, activation="relu"))
model1.add(Dropout(0.3))
model1.add(Dense(units=128, activation="relu"))
model1.add(Dropout(0.3))
model1.add(Dense(units=64, activation="relu"))
model1.add(Dropout(0.3))

# 출력층
model1.add(Dense(units=3, activation="softmax"))

model1.summary()
```

```python
model1.compile(loss="categorical_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
h1 = model1.fit(X_train, y_train_en, epochs=30, batch_size=30, validation_data=(X_test, y_test_en))
```

```python
# 시각화
import matplotlib.pyplot as plt

loss1 = h1.history["loss"]
val_loss1 = h1.history["val_loss"]
acc1 = h1.history["accuracy"]
val_acc1 = h1.history["val_accuracy"]

xaxis= range(1, 31)

plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
plt.plot(xaxis, loss1, "r", label="Train")
plt.plot(xaxis, val_loss1, "b", label="Test")
plt.legend()
plt.grid()
plt.title("LOSS")

plt.subplot(2, 1, 2)
plt.plot(xaxis, acc1, "r", label="Train")
plt.plot(xaxis, val_acc1, "b", label="Test")
plt.legend()
plt.grid()
plt.title("ACCURACY")
```

```python
# 예측하기
pred = model1.predict(X_test)

print(pred[0])
print(y_test_en.iloc[0])

print(pred[3])
print(y_test_en.iloc[3])

print(pred[30])
print(y_test_en.iloc[30])

print(pred[50])
print(y_test_en.iloc[50])
```

## 2. 소리 데이터를 Conv1D를 사용하여 학습

```python
# Conv1D에 입력하기 위해서 차원을 하나 증가
# 끝 인덱스에 차원 하나를 증가
X_train_ex = np.expand_dims(X_train, - 1)
X_test_ex = np.expand_dims(X_test, - 1)

X_train_ex.shape, X_test_ex.shape
```

### 신경망 설계

- kernel_size = 3  : 시간축 상에서 3개씩 끊는 것

<img src="https://drive.google.com/uc?id=139mNKnvsGOpsXmsLvR20jtKuBBqqCsb1">

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalAveragePooling1D
from tensorflow.keras.layers import Dense, Dropout, Flatten

model2 = Sequential()

# kernel_size = 3  : 시간축 상에서 3개씩 끊는 것
model2.add(Conv1D(filters=64, 
                  kernel_size=3,
                  input_shape=(30000, 1),
                  activation="relu"))
model2.add(Conv1D(filters=64,
                  kernel_size=3,
                  activation="relu"))
model2.add(MaxPooling1D(pool_size=3))

model2.add(Conv1D(filters=128,
                  kernel_size=3,
                  activation="relu"))
model2.add(Conv1D(filters=128,
                  kernel_size=3,
                  activation="relu"))
model2.add(GlobalAveragePooling1D())
model2.add(Dropout(0.5))

model2.add(Dense(units=3, activation="softmax"))

model2.summary()
```

```python
model2.compile(loss="categorical_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
h2 = model2.fit(X_train_ex, y_train_en, epochs=50, batch_size=30, validation_data=(X_test_ex, y_test_en))
```

```python
# 시각화
import matplotlib.pyplot as plt

loss2 = h2.history["loss"]
val_loss2 = h2.history["val_loss"]
acc2 = h2.history["accuracy"]
val_acc2 = h2.history["val_accuracy"]

xaxis= range(1, 31)

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(xaxis, loss2, "r", label="Train")
plt.plot(xaxis, val_loss2, "b", label="Test")
plt.legend()
plt.grid()
plt.title("LOSS")

plt.subplot(2, 1, 2)
plt.plot(xaxis, acc2, "r", label="Train")
plt.plot(xaxis, val_acc2, "b", label="Test")
plt.legend()
plt.grid()
plt.title("ACCURACY")
```

```python
# 예측하기
pred = model2.predict(X_test_ex)

print(pred[0])
print(y_test_en.iloc[0])

print(pred[3])
print(y_test_en.iloc[3])

print(pred[4])
print(y_test_en.iloc[4])

print(pred[30])
print(y_test_en.iloc[30])

print(pred[84])
print(y_test_en.iloc[84])
```

## 3. 사운드 데이터를 MFCC로 변환해서 Conv2D로 학습하기

```python
X_train = []
X_test = []
y_train = []
y_test = []
```

```python
# 훈련 데이터 변환
for fname in os.listdir(train_dir) :
  if ".wav" not in fname :
    continue
  
  label = fname.split(" ")[0]
  y_train.append(label)

  wav, sr = librosa.load(train_dir + fname)
  mfcc = librosa.feature.mfcc(wav)

  pad_x = pad2d(mfcc, 40)
  X_train.append(pad_x)
```

```python
# 테스트 데이터 변환
for fname in os.listdir(test_dir) :
  if ".wav" not in fname :
    continue
  
  label = fname.split(" ")[0]
  y_test.append(label)

  wav, sr = librosa.load(test_dir + fname)
  mfcc = librosa.feature.mfcc(wav)

  pad_x = pad2d(mfcc, 40)
  X_test.append(pad_x)
  
```

```python
# 리스트를 배열로 변환
X_train = np.array(X_train)
X_test = np.array(X_test)

y_train_en = pd.get_dummies(y_train)
y_test_en = pd.get_dummies(y_test)

X_train.shape, X_test.shape, y_train_en.shape, y_test_en.shape
```

```python
# 색상 차원 추가
X_train_ex = np.expand_dims(X_train, -1)
X_test_ex = np.expand_dims(X_test, -1)

X_train_ex.shape, X_test_ex.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dropout

model3 = Sequential()

model3.add(Conv2D(filters=64, 
                  input_shape=(20, 40, 1),
                  kernel_size=(3, 3),
                  activation="relu"))
model3.add(GlobalAveragePooling2D())  # Flatten 대신에 사용 (CNN층과 Dense층 사이에 사용)

model3.add(Dense(units=128, activation="relu"))
model3.add(Dropout(0.3))
model3.add(Dense(units=64,activation="relu"))
model3.add(Dropout(0.3))
model3.add(Dense(units=32, activation="relu"))
model3.add(Dropout(0.3))
model3.add(Dense(units="3", activation="softmax"))

model3.summary()
```

```python
model3.compile(loss="categorical_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
h3 = model3.fit(X_train_ex, y_train_en, epochs=100, batch_size=30, validation_data=(X_test_ex, y_test_en))
```

```python
# 시각화
import matplotlib.pyplot as plt

loss3 = h3.history["loss"]
val_loss3 = h3.history["val_loss"]
acc3 = h3.history["accuracy"]
val_acc3 = h3.history["val_accuracy"]

xaxis= range(1, 101)

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(xaxis, loss3, "r", label="Train")
plt.plot(xaxis, val_loss3, "b", label="Test")
plt.legend()
plt.grid()
plt.title("LOSS")

plt.subplot(2, 1, 2)
plt.plot(xaxis, acc3, "r", label="Train")
plt.plot(xaxis, val_acc3, "b", label="Test")
plt.legend()
plt.grid()
plt.title("ACCURACY")
```
