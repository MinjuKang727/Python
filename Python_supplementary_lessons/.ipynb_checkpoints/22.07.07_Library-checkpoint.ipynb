```python
import pandas as pd

#최대 읽어들이는 줄 수 옵션 지정(pandas Documentations에 더 많은 옵션 확인 가능)
pd.set_option('display.max_rows', 500)      # 최대 읽어 들이는 줄 수 : 500
pd.set_option('display.max_columns', 500)   # 최대 읽어 들이는 컬럼 수 : 500
pd.set_option('display.width', 100)   # 한 컬럼의 최대 너비 증가
```

```python
# 데이터 로딩
# CSV(Comma Seperated Values)
apt = pd.read_csv('./data/apt.csv', encoding = 'cp949')  # encoding은 utf-8로 default 되어 있음
print(apt.head(10))  # .head(숫자) :  맨 위쪽 열만 몇개 불러와 볼 수 있는 함수
print(apt.tail(10))  # .tail(숫자) :  맨 아래쪽 열만 몇개 불러와 볼 수 있는 함수
```

```python
apt.info()
```

### 1. 데이터 자르기
- 숫자 인덱스 or 인덱스 이름으로 자르는 함수
- iloc[행, 열]
- loc[행, 열]

```python
# 전체 데이터의 아파트 이름과 면적만 가져오기
print(apt.loc[:,'아파트':'면적'])
print(apt.iloc[ : , 4 : 6])
print(apt[['아파트','면적']])
```

```python
# 행 : 10000~20000, 열 : 건축년도, 도로명
print(apt.iloc[10000:20001,10:])
print(apt.iloc[10000:20001,-2:])
print(apt.loc[10000:20000, '건축년도':'도로명'])
print(apt.loc[10000:20000, '건축년도':])
print(apt[10000:20001][['건축년도','도로명']])
```

### 데이터 필터링(불리언 색인)

```python
print(apt['아파트'].unique())
```

```python
print(apt[(apt['아파트'] == '호반현대') & apt['층'] >= 5])
```

```python
# 아파트 면적이 130 이상인 것만 필터링
# 그중에 아파트 이름과 가격만 보고 싶다.
print(apt[apt['면적'] >= 130].loc[:,['아파트','가격']])
print(apt[apt['면적'] >= 130][['아파트','가격']])
```

```python
# 아파트 가격이 1.5억 미만인 아파트 이름과 가격을 필터링
print(apt[apt['가격'] < 15000][['아파트', '가격']])
```

## 4. 데이터 가공

```python
# 면적과 가격을 이용해서 단가 칼럼을 생성하고 추가해보자.
apt['단가'] = apt['가격'] / apt['면적']
```

```python
print(apt)
```

## 글자 가공

```python
print(apt['지역'].unique())
```

```python
def split_str(row) :
    return (row.split(" ")[0])
```

```python
print(apt['가격'].mean())
```

```python
apt['지역2'] = apt['지역'].apply(split_str)
```

```python
print(apt.head())
```

```python
# 지역2를 이용해서 광주광역시의 평균 아파트 가격을 알아보자.
print(apt[apt['지역2'] == '광주광역시']['가격'].mean())
```

## 5. 데이터 그룹핑

```python
apt[['지역2', '가격']].groupby(by = '지역2').mean()
```
